#Exercise 1: Markov Chain Story Generation

Instructions:
Write a program to generate a short story using Markov chains. Use a sample text (e.g., a
fairy tale) to build your Markov model. Start your generated story with the phrase:

"Once upon a time there was a kingdom..."

```python
story = """Once upon a time there was a peaceful kingdom surrounded by forests and rivers.
The kingdom was ruled by a wise king who cared deeply for his people.
In the village lived a young girl who dreamed of adventure beyond the castle walls.
One day a mysterious traveler arrived with stories of dragons and hidden treasure.
The people of the kingdom gathered to hear the tales and wondered if the stories were true.
Soon the young girl decided to explore the forest to discover the secrets of the world.
Along the journey she met a brave knight who protected travelers from danger.
Together they discovered that courage and kindness were the greatest treasures of all.
"""

# Create the training text file for the Markov chain
with open("fairytale.txt", "w", encoding="utf-8") as file:
    file.write(story)

print("fairytale.txt file created successfully and ready for Markov training.")
```

```python
import random
import re

# Load training text
with open("fairytale.txt", "r", encoding="utf-8") as file:
    text = file.read()

# Clean text
text = re.sub(r"[^\w\s]", "", text)
words = text.split()

# Build Markov chain
markov_chain = {}

for i in range(len(words) - 2):
    key = (words[i], words[i+1])
    next_word = words[i+2]

    if key not in markov_chain:
        markov_chain[key] = []

    markov_chain[key].append(next_word)
```

```python
# Starting phrase required by assignment
story = ["Once", "upon", "a", "time", "there", "was", "a", "kingdom"]

# Choose a random starting key from dataset
key = random.choice(list(markov_chain.keys()))

# Generate story
for _ in range(60):

    if key not in markov_chain:
        break

    next_word = random.choice(markov_chain[key])
    story.append(next_word)

    key = (key[1], next_word)

print("Generated Markov Story:\n")
print(" ".join(story))
```

#Exercise 2: LLM-Generated Story

The question is ‘Why Study Data Science Today?’

Instructions:
Use an LLM (e.g., OpenAI GPT, Hugging Face Transformers) to generate a creative story
about why it is important to study data science today.

```python
from transformers import pipeline

# Load text generation model
generator = pipeline("text-generation", model="gpt2")

prompt = """
Write a short creative story about a college student who discovers
why studying Data Science is important in today's world.
The story should include real-world impact such as healthcare,
business, or technology.
"""

story = generator(
    prompt,
    max_length=220,
    temperature=0.8,
    num_return_sequences=1
)

print("\nLLM Generated Story:\n")
print(story[0]["generated_text"])
```
