```python
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
#Reading the given file

import requests

url = "https://www.cnn.com/2025/06/13/style/why-luxury-brands-are-so-expensive"

headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url, headers=headers)
html_content = response.text

print("FIRST 700 CHARACTERS OF DOWNLOADED FILE:\n")
print(html_content[:700])
```

```python
from bs4 import BeautifulSoup
import json

soup = BeautifulSoup(html_content, "lxml")

article_text = ""

# find all json-ld scripts
scripts = soup.find_all("script", type="application/ld+json")

for script in scripts:
    try:
        data = json.loads(script.string)

        # Sometimes it's a list
        if isinstance(data, list):
            for item in data:
                if item.get("@type") == "NewsArticle":
                    article_text += item.get("articleBody", "")
        else:
            if data.get("@type") == "NewsArticle":
                article_text += data.get("articleBody", "")

    except:
        pass

print("FIRST 700 CHARACTERS OF ARTICLE TEXT:\n")
print(article_text[:700])
```

```python
#Remove HTML Tags

clean_text = BeautifulSoup(article_text, "html.parser").get_text()

print(clean_text[:700])
```

```python
#Lowercase + Remove Punctuation
import re

clean_text = article_text.lower()
clean_text = re.sub(r'[^\w\s]', '', clean_text)
```

```python
#Remove Stopwords
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
tokens = clean_text.split()

filtered_words = [word for word in tokens if word not in stop_words]
print(filtered_words)
```

```python
#Lemmatize
nltk.download('wordnet')
nltk.download('omw-1.4')

from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]

print("FIRST 50 LEMMATIZED WORDS:\n")
print(lemmatized_words[:50])
```

```python
#Compare With Stemming
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_words]

print("FIRST 50 STEMMED WORDS:\n")
print(stemmed_words[:50])
```
