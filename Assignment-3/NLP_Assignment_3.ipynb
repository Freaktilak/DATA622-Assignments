```python
from newspaper import Article

url = (
    "https://www.washingtonpost.com/world/2025/06/13/"
    "air-india-plane-crash-survivor-vishwash-kumar-ramesh/"
)

article = Article(url)
article.download()
article.parse()

full_text = article.text

# Load first 700 characters
text_700 = full_text[:700]

print("STEP 1: First 700 Characters of the Article\n")
print(text_700)
print(f"\nTotal characters loaded: {len(text_700)}")
```

```python
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
import nltk
nltk.download('punkt')

from nltk.tokenize import sent_tokenize

sentences = sent_tokenize(text_700)

print("\nSTEP 2: Sentences\n")
for i, sentence in enumerate(sentences):
    print(f"{i+1}. {sentence}")
print(f"\n Total sentences found: {len(sentences)}")
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

first_10_sentences = sentences[:10]

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(first_10_sentences)

# Convert to DataFrame for full display
tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=tfidf.get_feature_names_out()
)

print("STEP 4: FULL TF-IDF MATRIX\n")
print(tfidf_df)

print("\nTF-IDF Matrix Shape:", tfidf_matrix.shape)
```

```python
  embeddings = model.encode(sentences)

print("\nSTEP 5: Embedding of First Sentence\n")
print(embeddings[0])

print("\nEmbedding Shape:", embeddings[0].shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

similarity = cosine_similarity(
    [embeddings[0]],
    [embeddings[1]]
)

print("\nSTEP 6: Cosine Similarity Between Sentence 1 and 2\n")
print(similarity)
print("\nSimilarity Score:", similarity[0][0])
```
