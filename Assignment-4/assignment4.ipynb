```python
import spacy
import nltk
import benepar

# load spacy
nlp = spacy.load("en_core_web_sm")

# sentences from Gettysburg Address
sentence1 = """Four score and seven years ago our fathers brought forth on this continent, a new nation, conceived in Liberty, and dedicated to the proposition that all men are created equal."""

sentence2 = """Now we are engaged in a great civil war, testing whether that nation, or any nation so conceived and so dedicated, can long endure."""
```

Task 1 — Tokenization (spaCy + benepar)

```python
# spaCy tokenization
doc1 = nlp(sentence1)
doc2 = nlp(sentence2)

print("spaCy Tokens — Sentence 1:")
print([token.text for token in doc1])

print("\nspaCy Tokens — Sentence 2:")
print([token.text for token in doc2])


# benepar (via nltk tokenizer)
tokens1 = nltk.word_tokenize(sentence1)
tokens2 = nltk.word_tokenize(sentence2)

print("\nbenepar/NLTK Tokens — Sentence 1:")
print(tokens1)

print("\nbenepar/NLTK Tokens — Sentence 2:")
print(tokens2)
```

Task 2 — POS Tagging (Sentence 1)

```python
doc = nlp(sentence1)

print("Token\t\tPOS\t\tDetailed Tag")
print("---------------------------------------")

for token in doc:
    print(f"{token.text:<12}{token.pos_:<12}{token.tag_}")
```

Task 3 — Dependency Parsing (Sentence 2)

```python
print("Token\t\tDependency\tHead Word")
print("------------------------------------------------")

for token in doc:
    print(f"{token.text:<12}{token.dep_:<15}{token.head.text}")
```

Task 4 — Constituency Parsing (NLTK + benepar)

```python
import benepar
import nltk

parser = benepar.Parser("benepar_en3")

tokens = nltk.word_tokenize(sentence1)

tree = parser.parse(tokens)

print(tree)
```

```python
tree.pretty_print()
```

Task 5 — Extract Noun Phrases

```python
print("Noun Phrases — Sentence 1:\n")
for chunk in doc1.noun_chunks:
    print(chunk.text)

print("\nNoun Phrases — Sentence 2:\n")
for chunk in doc2.noun_chunks:
    print(chunk.text)
```

Task 6 — CRF vs HMM

HMM is a generative sequence model that assumes independent observations.
CRF is a discriminative model that considers full contextual features and usually gives higher accuracy for sequence labeling tasks like POS tagging and Named Entity Recognition.
